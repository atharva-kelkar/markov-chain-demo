```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from time import sleep
from IPython.display import clear_output
```

```python
df = pd.read_excel('mc_probabilities.xlsx', header=0, index_col=0)
probs = df.to_numpy()
```

```python
cities = ['Averagemont', 'Bayesville', 'Continuopolis', 'Discretown', 'East Vandermonde']
```

```python
from random import choices

class MarkovChain():
    
    def __init__(self, cities, probs):
        self.cities = np.array( cities )
        self.probs = probs
        
        if len(self.cities) != len(probs):
            print('Length of cities and probabilities should be equal!')
        
    def get_total_count(self, subset_cities):
        total_count = np.zeros(len(self.cities))
        
        count = 0
        for city in self.cities:
            total_count[count] = np.sum( np.array(subset_cities) == city )
            count += 1
        
        return total_count / total_count.sum()
        
#     def live_plot(self):
        
        
    def add_city(self, curr_city):
        next_city = choices( self.cities, self.probs[ self.cities == curr_city, : ].squeeze() )
        self.cities_all.append( next_city[0] )
        
    def run_simulation(self, init_city, N_trips):
        assert (init_city in self.cities), 'City mentioned is not in cities listed!'
        
        ## Define current city as initial city for starting off
        self.cities_all = [ init_city ]

        for i in range(N_trips):
            ## Add next city
            self.add_city( self.cities_all[-1] )
        
        ## Convert list to numpy array
        self.cities_all = np.array( self.cities_all )
        
        
    def live_plot(self, i):
        plt.cla()
#         clear_output(wait=True)
        subset_cities = self.cities_all[ : (i+1) ]
        total_count = self.get_total_count( subset_cities )
        self.ax.bar( self.x, total_count, align='center' )
        self.ax.set_xticks( self.x, self.cities)
        self.ax.set_ylim([0, 1.0])
        self.ax.set_xlabel('City', fontsize=14)
        self.ax.set_ylabel('Ratio of time spent', fontsize=14)
        
#         return self.ax
            
    def plot_simulation(self, N_trips):
        
        self.x = np.arange(len(self.cities))
        
        self.fig, self.ax = plt.subplots()    
        

        self.ani = FuncAnimation( self.fig, self.live_plot, 
                                 frames=N_trips, interval=10 )
#             sleep(0.001)

    def run_and_plot_simulation(self, init_city, N_trips, to_plot=True ):
        ## Run simulation first
        self.run_simulation(init_city, N_trips)

        ## Plot simulation
        self.plot_simulation(N_trips)
        
        plt.show()
```

```python
%matplotlib inline

mc = MarkovChain( cities, probs )
mc.run_and_plot_simulation(cities[2], 100)
```
